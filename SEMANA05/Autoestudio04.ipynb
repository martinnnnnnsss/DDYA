**Defina** una clase `Punto2D` que representa un punto en un espacio bidimensional.


*   **Atributos:** La posición ($x,y$)
*   **Métodos:** (i) Mover el punto una distancia $dx$ en horizontal y una distancia $dy$ en vertical.
(ii) Calcular la distancia del punto al centro ($0,0$)



image.png

````python
# Punto2D
class Punto2D:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def mover(self,dx,dy):
    self.x += dx
    self.y += dy

  def distancia_a_centro(self):
    return  math.sqrt(self.x**2 + self.y**2)
````

# Stacks (pilas)

Ahora, tomando lo anterior como base, vamos a definir nuestras estructuras de datos usando programacion orientada a objetos (OOP).

Las pilas son estructuras de datos estaticas que poseen operaciones de insercion y borrado, donde para el caso de una pila, el elemento borrado es el elemento que fue insertado mas recientemente (LIFO).

Definamos una clase `Stack` que represente una pila como la vista en clase.

````python
class Stack:
  def __init__(self, n: int):
    assert n > 0
    self.A: list = [None for _ in range(n)]
    self.top = -1

  def is_empty(self) -> bool:
    return self.top == -1

  def push(self, v):
    assert not self.is_full()
    self.top += 1
    self.A[self.top] = v

  def pop(self):
    assert not self.is_empty()
    v = self.A[self.top]
    self.top -= 1
    return v

  def is_full(self) -> bool:
    return self.top == len(self.A) -1

  def __repr__(self) -> str:
    values = ", ".join(str(self.A[i]) for i in range(0, self.top+1))
    return f"[{values}]"
````

````python
# Prueba trivial
n = 10
s = Stack(n)
for i in range(n):
  s.push(i)

while not s.is_empty():
  print(s.pop(), end=" ")
````

### Problema: balanceo correcto

**Problema**

Una cadena de texto con paréntesis `()` y `[]` se le considerada _correcta_ si los paréntesis están cerrados de manera adecuada. Es decir, una cadena es _correcta_ si.
   * Es una cadena de texto vacia, o
   * si $A$ y $B$ son correctas, entonces $AB$ es correcta, o
   * si $A$ es correcta, entonces $(A)$ y $[A]$ son correctas

Implementemos una solucion que determine si una cadena de texto de paréntesis es _correcta_.

**Solucion**

Para esto podemos notar que para el balanceo de paréntesis es necesario que los paréntesis esten en un mismo 'nivel', una forma de pensarlo es visualizandolo como un 'pastel'.

Tomemos como ejemplo `[([[]])]`, este puede visualizarse como
```
   []    <- cuarto nivel
  [  ]   <- tercer nivel
 (    )  <- segundo nivel
[      ] <- primer nivel
```

O tambien. `([])[([])]` se puede visualizar como
```
       []
 []   (  )
(  ) [    ]
```

Un caso como `[[([))]` no esta balanceado, si lo vieramos visualmente quedaria
```
   [)    <- No coinciden los paréntesis
  (  )
 [    ]
[        <- No tiene paréntesis que lo empareje
```

Cada nivel se puede representar dentro de una pila como un elemento, de manera que cuando procesamos un paréntesis abierto, sea `(` o `[` estamos procesando un nuevo nivel, por tanto lo _apilamos_ en la pila, y solo cuando encontramos un paréntesis cerrado `]` o `)` es que decidimos cerrar un nivel, por tanto _desapilamos_ de la pila.

Para saber que esta balanceada tenemos tres casos:
* Si hay un paréntesis abierto, debemos guardarlo en la pila.
* Si hay un paréntesis cerrado, a pila debe tener al menos un elemento apilado y el elemento en la parte superior de la pila debe ser el paréntesis correspondiente
* Si procesamos todos los paréntesis, la pila al final debe estar vacia

````python
  def parenthesis_balance(pars: str) -> bool:
    size = len(pars)
    s = Stack(size+1)
    #La pila no puede ser de tamaño cero
    for par in pars:
      if par in "([":
        # Si es un paréntesis abierto, apilamos a la pila
        s.push(par)
      else: # Tenemos un paréntesis cerrado
        if s.is_empty():
          # Si la pila esta vacia, entonces no estamos balanceados
          return False
        # Sino, tenemos un paréntesis cerrado y comprobamos la primera condicion
        e = s.pop()
        if e == "(" and par != ")":
          # Si tenemos (], entonces no esta balanceado
          return False
        if e == "[" and par != "]":
          # Si tenemos [), entonces no esta balanceado tampoco
          return False
    # La pila debe estar vacia al finalizar
    return s.is_empty()
````

````python
# Comprobamos de que funcione con diferentes casos
assert parenthesis_balance("") == True, "la cadena vacia está balanceada"
assert parenthesis_balance("[]()") == True, "[]() está balanceada"
assert parenthesis_balance("[()]") == True, "[()] está balanceada"
assert parenthesis_balance("[(])") == False, "[(]) no está balanceada"
assert parenthesis_balance("])") == False, "]) no está balanceada"
````

### $\color{blue}{Ejercicio: calculadora}$

Una calculadora posfija realiza calculos en [notacion polaca inversa (RPN por sus siglas en ingles)](https://es.wikipedia.org/wiki/Notaci%C3%B3n_polaca_inversa), esto es que las expresiones se escriben de manera que los operandos se escriben primero que los simbolos de los operadores, tome como ejemplo:
     * `2 3 +` es equivalente a `2 + 3` (en notacion _infija_)
     * `3 4 5 + *` es equivalente a `3 * (4 + 5)`
     * `5 1 2 + 4 * + 3 -` es equivalente a `5 + ((1 + 2) * 4) - 3`

**Implemente** una calculadora posfija que reciba los argumentos en notacion polaca inversa por medio de usar la clase `Stack` que soporte los operadores `+`, `-`, `*` y `/`.

> **_Ayuda_**: el tipo `str` tiene un metodo llamado `isdigit` que responde si una cadena de texto representa un numero, puede usar este metodo para determinar facilmente si el elemento que se esta procesando es un numero o uno de los operadores. De manera que `"15".isdigit() == True`, y `"noesnumero".isdigit() == False`

````python
class Stack:
  def __init__(self, n: int):
    assert n > 0
    self.A: list = [None for _ in range(n)]
    self.top = -1

  def is_empty(self) -> bool:
    return self.top == -1

  def push(self, v):
    assert not self.is_full()
    self.top += 1
    self.A[self.top] = v

  def pop(self):
    assert not self.is_empty()
    v = self.A[self.top]
    self.top -= 1
    return v

  def is_full(self) -> bool:
    return self.top == len(self.A) -1

  def __repr__(self) -> str:
    values = ", ".join(str(self.A[i]) for i in range(0, self.top+1))
    return f"[{values}]"
def rpn(expression: str) -> int:
    stack = Stack(20)
    args: list[str] = expression.split()
    # Implemente
    for x in args:
        if x.isdigit():
            stack.push(int(x))
        else:
            b = stack.pop()
            a = stack.pop()

            if x == '+':
                stack.push(a + b)
            elif x == '-':
                stack.push(a - b)
            elif x == '*':
                stack.push(a * b)
            elif x == '/':
                stack.push(a / b)
    return stack.pop()
````

````python
assert rpn("2 3 +") == 5
assert rpn("3 4 5 + *") == 27
assert rpn("5 1 2 + 4 * + 3 -") == 14
````

### $\color{blue}{Ejercicio: full, enqueue, dequeue}$

````python
from typing import Any # Para denotar que es cualquier tipo de datos

class Queue:
  def __init__(self, n: int):
    assert n > 0
    self.A: list[Any] = [None for _ in range(n+1)]
    self.head: int = 0
    self.tail: int = 0

  def is_empty(self) -> bool:
    return self.head == self.tail

  def is_full(self) -> bool: # Implemente
    return (self.tail+1) % len(self.A) == self.head

  def enqueue(self, val: Any): # Implemente
    assert not self.is_full()
    self.A[self.tail] = val
    self.tail = (self.tail+1) % len(self.A)

  def dequeue(self) -> Any:
    assert not self.is_empty()
    # Implemente
    val = self.A[self.head]
    self.head = (self.head+1) % len(self.A)
    return val
````

````python
# Prueba trivial
n = 10
q = Queue(n)
for i in range(n):
  q.enqueue(i)

while not q.is_empty():
  print(q.dequeue(), end=" ")
````

# Linked List (listas encadenadas)

Una lista encadenada es una estructura de datos en la cual los objetos estan organizados en un orden lineal. A diferencia de un arreglo, en la que los objetos almacenados se pueden acceder por medio de indices, el orden en una lista encadenada esta determinada por *punteros* en cada objeto.

Para esta, necesitamos de dos clases, particularmente `Node`, cada uno de los nodos de nuestra lista encadenada, y `LinkedList`, que contiene el apuntador al primer nodo de la lista.

Definamos la clase `Node` con los atributos que necesitamos.

##Nodo

````python
from typing import Any

class Node:
  def __init__(self, key: Any):
    self.key: Any = key
    self.next: Node | None = None
    self.prev: Node | None = None

  def __repr__(self) -> str:
    return f"Node[{self.key}]"
````

> Si se pregunta que significa la sintaxis `v: Node | None`, esto significa que el valor que guarda la variable `v` puede tener el valor de un nodo tipo `Node`, pero tambien puede contener el valor de `None`
>
> Como caso de ejemplo, si tuvieramos una variable que tiene como tipo `v: int | str`, significa que lo que guarda la variable `v` puede ser un entero `int` o una cadena de texto `str`.

##Insertar un nodo
Adicionalmente, definamos la funcion de ayuda `insert_node(z: Node, x: Node)`, cuya labor es insertar un nuevo nodo entre dos nodos, donde `x` es el nuevo nodo a insertar, y `y` el objeto que ya hace parte de una lista encadenada.

El resultado de la operacion sera que dada la lista `[x, y]` con un nuevo nodo `z`, si se ejecuta `insert_node(z, x)` se tendra como resultado que la lista ahora es `[x, z, y]`.

Este nos permite agregar un nodo en nuestra lista en una posicion arbitraria en tiempo $\Theta(1)$.

````python
def insert_node(z: Node, x: Node):
  assert z is not None
  assert x is not None
  z.next = x.next
  z.prev = x
  if x.next is not None:
    x.next.prev = z
  x.next = z
````

## `LinkedList`
Ahora definamos la clase `LinkedList` para nuestra lista encadenada.

````python
class LinkedList:
  def __init__(self):
    self.head: Node | None = None

  def prepend(self, key: Any):
    x = Node(key)
    x.next = self.head
    x.prev = None
    if self.head is not None:
      self.head.prev = x
    self.head = x

  def search(self, k: Any) -> Node | None:
    x = self.head
    while x is not None and x.key != k:
      x = x.next
    return x

  def delete(self, x: Node):
    assert x is not None
    if x.prev is not None:
      x.prev.next = x.next
    else:
      self.head = x.next

    if x.next is not None:
      x.next.prev = x.prev

  def __repr__(self) -> str:
    # Debemos ir por cada elemento para poder imprimirlo
    values = []
    x = self.head
    while x is not None:
      values.append(x)
      x = x.next
    return str(values)
````

Esta implementacion es la misma vista en el libro guia, si probamos los metodos, veremos lo siguiente:

````python
ll = LinkedList()
for i in range(10):
  ll.prepend(i)

print(ll)
s = ll.search(3)
assert s is not None
print(s) # Debe ser el nodo que contiene el numero 3
insert_node(Node(15), s)
print(ll.search(15)) # Debe ser el nuevo nodo que agregamos a la lista que contiene el numero 15
print(ll.search(100)) # Debe ser None por no estar en la lista

# Luego para borrar un elemento de la lista, primero debemos buscar el nodo
node = ll.search(4)
assert node is not None
print("A borrar", node)
ll.delete(node)
print(ll)
````

Notese que `insert_node` no hace parte de nuestra clase `LinkedList`, esto por que el metodo solo opera sobre dos nodos arbitrarios, sin necesidad de acceder al unico atributo interno de la lista (`head`).

## $\color{blue}{Ejercicio. invert}$

**Realice** una extensión a `LinkedList` para incluir una funcion llamada `invert`, que invierta la direccion de la lista en tiempo $\Theta(n)$.

````python
  def invert(self:LinkedList):
    current = self.head
    prev_node = None
    while current is not None:
        next_node = current.next
        current.next = prev_node
        current.prev = next_node
        prev_node = current
        current = next_node
    self.head = prev_node
  LinkedList.invert=invert
````

````python
casos = [
    [],
    [1],
    [1, 2, 3],
    [10, 20, 30, 40, 50]
]

for datos in casos:
    lista = LinkedList()

    # Cargar los datos en el orden correcto
    for x in reversed(datos):
        lista.prepend(x)

    print(f"Original: {lista}")
    lista.invert()
    print(f"Invertida: {lista}\n")
````

## $\color{blue}{Ejercicio. remove\_duplicates}$
**Realice** una modificacion sobre `LinkedList` para incluir una funcion llamada `remove_duplicates`, que remueva los elementos repetidos de la lista en tiempo $\Theta(n^2)$.

````python
  def remove_duplicates(self: LinkedList):
    current = self.head
    while current is not None:
        dupli = current.next
        while dupli is not None:
            next_node = dupli.next  # Guardamos el siguiente por si borramos
            if dupli.key == current.key:
                self.delete(dupli)
            dupli = next_node
        current = current.next

LinkedList.remove_duplicates = remove_duplicates

````

````python
lista = LinkedList()
listaaa = [3, 5, 3, 2, 5, 7, 2]
for x in listaaa:
    lista.prepend(x)

print("Antes de eliminar duplicados:")
print(listaaa)

lista.remove_duplicates()

print("Después de eliminar duplicados:")
print(lista)
````

## $\color{blue}{Ejercicio. concat}$
**Realice** una extensión `LinkedList` para incluir una funcion llamada `concat`, que haga union de dos listas encadenadas en tiempo $\Theta(1)$.

````python
 def concat(self:LinkedList, other: LinkedList): # <<<< Implemente
    if self.head is None:
        self.head = other.head
        return

    if other.head is None:
        return
    last = self.head
    while last.next is not None:
        last = last.next
    last.next = other.head
    other.head.prev = last
LinkedList.concat=concat #adicionamos el nuevo método a la clase LinkedList
````
